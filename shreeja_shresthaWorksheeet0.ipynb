Problem - 4: Matrix Operations:

```python
import numpy as np

#Given matrices
A = np.array([[3, 4],
              [7, 8]])

B = np.array([[5, 3],
              [2, 1]])
```

```python
# 1.Prove A.A⁻¹ = I

A_inv = np.linalg.inv(A)
result = np.dot(A, A_inv)

print("A inverse:\n", A_inv)
print("A * A_inv:\n", result)
```

```python
# 2.Prove AB ≠ BA

AB = np.dot(A, B)
BA = np.dot(B, A)

print("AB:\n", AB)
print("BA:\n", BA)
```

```python
# 3.Prove (AB)^T = B^T A^T

AB_T = np.dot(A, B).T
BT_AT = np.dot(B.T, A.T)

print("(AB)^T:\n", AB_T)
print("B^T A^T:\n", BT_AT)
```


1. Element-wise Addition:
• Using Python Lists, perform element-wise addition of two lists of size 1, 000, 000. Measure
and Print the time taken for this operation.
• Using Numpy Arrays, Repeat the calculation and measure and print the time taken for
this operation.



```python
import time
import numpy as np

size = 1000000

# Create two Python lists
list1 = list(range(size))
list2 = list(range(size))

# Element-wise addition using Python lists
start = time.time()

result_list = [list1[i] + list2[i] for i in range(size)]

end = time.time()
print("Time using Python lists:", end - start, "seconds")


# Convert lists to NumPy arrays
array1 = np.array(list1)
array2 = np.array(list2)

# Element-wise addition using NumPy
start = time.time()
result_array = array1 + array2

end = time.time()
print("Time using NumPy arrays:", end - start, "seconds")
```

3. Dot Product
• Using Python Lists, compute the dot product of two lists of size 1, 000, 000. Measure and
Print the time taken for this operation.
• Using Numpy Arrays, Repeat the calculation and measure and print the time taken for
this operation.

```python
import time
import numpy as np

size = 1000000

#Create two Python lists
list1 = list(range(size))
list2 = list(range(size))

#Dot product using Python lists
start = time.time()

dot_product_list = sum(list1[i] * list2[i] for i in range(size))

end = time.time()
print("Dot product using Python lists:", dot_product_list)
print("Time using Python lists:", end - start, "seconds")


#Convert lists to NumPy arrays
array1 = np.array(list1)
array2 = np.array(list2)

#Dot product using NumPy
start = time.time()

dot_product_numpy = np.dot(array1, array2)

end = time.time()
print("Dot product using NumPy:", dot_product_numpy)
print("Time using NumPy arrays:", end - start, "seconds")
```

4. Matrix Multiplication

• Using Python lists, perform matrix multiplication of two matrices of size 1000x1000. Mea-
sure and print the time taken for this operation.

• Using NumPy arrays, perform matrix multiplication of two matrices of size 1000x1000.
Measure and print the time taken for this operation.

```python
import time
import numpy as np
import random

size = 1000

# Create two 1000x1000 matrices using Python lists
matrix1 = [[random.randint(1,10) for _ in range(size)] for _ in range(size)]
matrix2 = [[random.randint(1,10) for _ in range(size)] for _ in range(size)]

# Matrix multiplication using Python lists
start = time.time()

result = [[0]*size for _ in range(size)]

for i in range(size):
    for j in range(size):
        for k in range(size):
            result[i][j] += matrix1[i][k] * matrix2[k][j]

end = time.time()

print("Time taken using Python lists:", end - start, "seconds")


# Convert to NumPy arrays
array1 = np.array(matrix1)
array2 = np.array(matrix2)

# Matrix multiplication using NumPy
start = time.time()

result_numpy = np.dot(array1, array2)

end = time.time()

print("Time taken using NumPy arrays:", end - start, "seconds")
```
